## Tokenization

```python
#importing nltk(Natural Language Toolkit)
import nltk

#Downloading Package punkt
nltk.download('punkt')
```

```python
"""
Tokenization is the process of breaking a stream of text up into words, phrases, symbols, or other meaningful elements called tokens
"""
```

```python
#opening file in read mode
file = open('input.txt', 'r')

#Reading the data in the file
input = file.read()
```

```python
#Tokenizing into Sentences, Words

stokens = nltk.sent_tokenize(input) #Sentence Tokens
wtokens = nltk.word_tokenize(input) #Words Tokens
```

```python
print("Sentence Tokens:", stokens)
```

```python
print("Word Tokens:", wtokens)
```

## POS(Part of Speech tagging)

```python
"""
The process of classifying the words in a text(corpus) into their parts of speech and labeling them accordingly is known as part-of-speech tagging
"""
```

```python
#Downloading required packages
nltk.download('averaged_perceptron_tagger')
```

```python
#performing Part of Speech tagging using pos_tag
pos=nltk.pos_tag(wtokens)
```

```python
print("Part of Speech tagging:",pos)
```

## Stemming

```python
"""
Stemming is the process for reducing injected words to their stem, base root form
"""
```

```python
#Importing required pacakges
from nltk.stem import PorterStemmer
from nltk.stem import LancasterStemmer
from nltk.stem import SnowballStemmer
```

```python
#PorterStemmer
pStemmer = PorterStemmer()
print("Porter Stemmer:")
for i in wtokens:
    print(pStemmer.stem(i), end='')
```

```python
#LancasterStemmer
lStemmer= LancasterStemmer()
print("Lancaster Stemmer:")
for i in wtokens:
    print(lStemmer.stem(i), end='')
```

```python
#SnowballStemmer
sStemmer= SnowballStemmer('english')
print("Snowball Stemmer:")
for i in wtokens:
    print(sStemmer.stem(i), end='')
```

## Lemmatization 

```python
"""
Lemmatization process involves first determining the part of speech of a word and applying different normalization rules for each part of speech
"""
```

```python
#importing required packages
from nltk.stem import WordNetLemmatizer

#Downloading required packages
nltk.download('wordnet')
```

```python
#Applying Lemmatization using WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
print('Lemmatization:')
for i in wtokens:
    print(lemmatizer.lemmatize(str(i)),end=' ')
```

## Trigram

```python
"""
An n-gram is a contiguous sequence of n items from a given sample of text or speech(for Trigram n=3)
"""
```

```python
#Importing required packages
from nltk import ne_chunk, ngrams
```

```python
print("Trigrams:")
trigrams = list(ngrams(wtokens,3)) #In N_gram n would be 3 for Trigram 
trigrams
```

## Named Entity Recognition

```python
#Importing required packages
from nltk.tokenize import wordpunct_tokenize
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
```
