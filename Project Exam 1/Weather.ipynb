### Removing outlier data

```python
#Removing the outlier data using the z-score for the complete dataframe

from scipy import stats

"""
 Z-score is measured in terms of standard deviations from the mean. If a Z-score is 0, it indicates that the data point's score is identical to the mean score. A Z-score of 1.0 would indicate a value that is one standard deviation from the mean. Z-scores may be positive or negative, with a positive value indicating the score is above the mean and a negative score indicating it is below the mean.
"""

z = np.abs(stats.zscore(df))
```

```python
#let us consider threshold is 3

# we shall remove all the points which are 3 units away standard deviation from the mean 
np.where(z < 3)
```

```python
#Filtering the Dataframe with the values which are 3 units away standard deviation from the mean
filtered_entries = df[(z < 3).all(axis=1)]
```

```python
#plotting the scatter plot with the filtered dataframe & labeling x,y and title
plt.scatter(filtered_entries["Apparent Temperature (C)"],filtered_entries["Temperature (C)"])
plt.xlabel('Apparent Temperature (C)')
plt.ylabel('Temperature (C)')
plt.title('After removing outliers');
```

```python
#Feature Matrix Creation
X=filtered_entries[['Apparent Temperature (C)','Daily Summary','Precip Type','Humidity',]]
```

```python
X
```

```python
#Label Matrix
y=filtered_entries['Temperature (C)']
```

```python
#importing packages to split into train & test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
#No. of samples in test dataframe
len(X_test)
```

```python
#No. of samples in test dataframe
len(y_test)
```

```python
#Histogram to know frequency of data
plt.hist(y_train);
```

```python
#Skewness of the data
y_train.skew()
```

### Evaluate the model

```python
#Importing the packages for LinearRegression
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X_train,y_train) #Fit linear model
y_pred=reg.predict(X_test) #Predict using the linear model
```

```python
#No. of samples in test dataframe
len(y_pred)
```

### interpreting the performance

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
r2_score = r2_score(y_test,y_pred) #Calculating r2 score
print("r2 Score:",r2_score)
rmse=mean_squared_error(y_test,y_pred) #calculating rmse(root mean squared error)
print("rmse score:",rmse)
```

r2 score represents how well the model fits the data. Here R2 score is 0.98 which is close to 1 which means regression model fits observation good. R-squared is a relative measure of fit.

RMSE is how close the observed data points are to the model's predicted values. RMSE is an absolute measure of fit. Lower values of RMSE indicate better fit. But we have it as 1.08, this is due to:

There are many factors that may have contributed to this inaccuracy, for example :

Need more data: We need to have a huge amount of data to get the best possible prediction.

Bad assumptions: We made the assumption that this data has a linear relationship, but that might not be the case. Visualizing the data may help you determine that.

Poor features: The features we used may not have had a high enough correlation to the values we were trying to predict.
