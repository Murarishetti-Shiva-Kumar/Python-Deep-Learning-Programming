### Applying Tokenization, Part of speech tagging, stemming, Lemmetization

```python
# finding null values
dataset.isna().sum()
```

```python
#Creating Feature dataframe
X=dataset['Text']
```

```python
#Creating Target Dataframe
y=dataset['Class']
```

### Tokenization

```python
# importing Natural Language Toolkit package
import nltk
```

```python
# Word Tokenization
wtokens=[]
for dur in X:
    tokens=nltk.word_tokenize(str(dur))
    wtokens.append(tokens)
```

```python
print(wtokens,end='')
```

```python
# word Tokenization
wtokens_test=[]
for dur in X:
    tokens=nltk.word_tokenize(str(dur))
    words = [word for word in tokens if word.isalpha()] # removing special characters
    wtokens_test.append(words)
```

```python
print(wtokens_test,end='')
```

```python
# sentence Tokenization
stokens=[]
for dur in X:
    tokens=nltk.sent_tokenize(str(dur))
    stokens.append(tokens)
```

```python
stokens
```

### Part of Speech tagging

```python
# Part of Speech tagging
pos=nltk.pos_tag(str(wtokens_test))
```

```python
pos
```

### Stemming

```python
# importing packages for stemming
from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer
```

```python
# porter Stemmer
pStemmer=PorterStemmer()
pStemme=[]
for dur in X:
    token=pStemmer.stem(dur)
    pStemme.append(token)
```

```python
pStemme
```

```python
# Lancaster Stemmer
lStemmer=LancasterStemmer()
lStemme=[]
for dur in X:
    token=lStemmer.stem(dur)
    lStemme.append(token)
```

```python
lStemme
```

```python
# Snowball Stemmer
sStemmer=SnowballStemmer('english')
sStemme=[]
for dur in X:
    token=sStemmer.stem(dur)
    sStemme.append(token)
```

```python
sStemme
```

### Lemmetization

```python
# importing packages for wordnet Lemmatizer
from nltk.stem import WordNetLemmatizer
```

```python
# Wordnet Lemmatizer
lemmatizer=WordNetLemmatizer()
lemmatize=[]
for dur in X:
    token=lemmatizer.lemmatize(dur)
    lemmatize.append(token)
```

```python
lemmatize
```

### Named Entity Recognition

```python
# importing packages for finding Named Entity Recognition (NER)
from nltk import ne_chunk, wordpunct_tokenize, pos_tag
```

```python
# calculating NER
ner=ne_chunk(pos_tag(wordpunct_tokenize(str(X))))
```

```python
ner
```

### CountVectorizer

```python
# importing package for count vectorization
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Importing packages for spltiing into train and test
from sklearn.model_selection import train_test_split
```

```python
# splitting into train and test data with 20% as test size
X_train,X_test,y_train,y_test=train_test_split(lemmatize,y,test_size=0.2) #passing the data after applying lemmetization
```

```python
#CountVectorizer Converts collection of text documents to a matrix of token counts
count_vector = CountVectorizer(stop_words='english') #applying 'stop words' english
X_train_counts=count_vector.fit_transform(X_train,y_train) #Learn the vocabulary dictionary and return document-term matrix
```

```python
# importing Random forest package
from sklearn.ensemble import RandomForestClassifier
```

```python
# Applying Random forest
clf = RandomForestClassifier() 
clf.fit(X_train_counts,y_train) #Build a forest of trees from the training set (X, y)
```

```python
X_test_counts=count_vector.transform(X_test) #Transform documents to document-term matrix
y_pred=clf.predict(X_test_counts) #Perform classification on samples in X
```

```python
# importing packages for finding accuracy score
from sklearn.metrics import accuracy_score
score_count=accuracy_score(y_test, y_pred) #Accuracy Score
score_count
```

### TfidfVectorizer

```python
#TF(t) = (Number of times term t appears in a document) / (Total number of terms in the document)
#IDF(t) = log_e(Total number of documents / Number of documents with term t in it)

#importing TfidfVectorizer package
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vector=TfidfVectorizer(stop_words='english')#applying 'stop words' english
X_train_tfidf=tfidf_vector.fit_transform(X_train,y_train) #Learn vocabulary and idf, return document-term matrix
```

```python
# Applying Random forest
clf = RandomForestClassifier()
clf.fit(X_train_tfidf,y_train) #Build a forest of trees from the training set (X, y)
```

```python
X_test_tfidf=tfidf_vector.transform(X_test) #Transform documents to document-term matrix
y_pred=clf.predict(X_test_tfidf) #Perform classification on samples in X
```

```python
score_tfidf=accuracy_score(y_test, y_pred) #Accuracy Score
score_tfidf
```

### Evaluate the model and try to interpret the result

```python
print("Accuracy Score for Count Vectorizer:", score_count)
print("Accuracy Score for Tfidf Vectorizer:", score_tfidf)
```
